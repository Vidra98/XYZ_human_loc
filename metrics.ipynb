```python
import scipy.io
import cv2
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# Midas small metrics scaled the full GT

```python
to_plot=pd.read_csv('metrics/metrics_midas_small_4m_scaled.csv')

#plt.figure()
RMS_midas_small_zoomed_plt = plt.figure()
sns.set_theme(style="whitegrid")
ax = sns.boxplot(data=to_plot,x="file number",y="RMS")
ax.set(xlabel='file number', ylabel='Root mean square error [meter]')
ax.set(ylim=(0, 3))

abs_rel_plt=plt.figure()
sns.set_theme(style="whitegrid")
ax = sns.boxplot(data=to_plot,x="file number",y="abs rel")
ax.set(xlabel='file number', ylabel='Absolute relative')
ax.set(ylim=(-1, 1))

```
