```python
import scipy.io
import cv2
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
to_plot=pd.read_csv('metrics_midas_small_4m_scaled.csv')

#plt.figure()
RMS_midas_small_plt = plt.figure()
sns.set_theme(style="whitegrid")
ax = sns.boxplot(data=to_plot,x="file number",y="RMS")
ax.set(xlabel='file number', ylabel='Root mean square error [meter]')
ax.set(ylim=(0, 4.5))
plt.savefig('Figures/RMS_midas_small_scaled.eps')

RMS_midas_small_zoomed_plt = plt.figure()
sns.set_theme(style="whitegrid")
ax = sns.boxplot(data=to_plot,x="file number",y="RMS")
ax.set(xlabel='file number', ylabel='Root mean square error [meter]')
ax.set(ylim=(0, 2))
plt.savefig('Figures/RMS_midas_small_scaled_zoomed.eps')

abs_rel_plt=plt.figure()
sns.set_theme(style="whitegrid")
ax = sns.boxplot(data=to_plot,x="file number",y="abs rel")
ax.set(xlabel='file number', ylabel='Absolute relative')
ax.set(ylim=(-1, 1))
plt.savefig('Figures/ABS_REL_midas_small_scaled.eps')

plt.figure()
sns.relplot(
    data=to_plot, kind="line",
    y="mean pred", x="mean GT",
    hue="file number"
)
```

```python
to_plot=pd.read_csv('divers/metrics_midas_3m_no_scaling.csv')

error=to_plot["Error"].values
error[error == 0] = np.nan
print('For a range of 0 to 3m :\n',error)
print('The RMS error of rescaled data is :',np.nanmean(np.abs(error)),'with a std of :',np.nanstd(np.abs(error)))
print('The absolute error of rescaled data is :',np.nanmean((error)),'with a std of :',np.nanstd((error)))
```
